```python
python_df = python_df.rename(columns={
    'Country' : 'NumRespondents',
    'LanguageWorkedWith' : 'PeopleKnowPython'
} )

python_df
```

```python
python_df['PercentagePeopleKnowPython'] = (python_df['PeopleKnowPython'] / python_df['NumRespondents']) * 100
```
